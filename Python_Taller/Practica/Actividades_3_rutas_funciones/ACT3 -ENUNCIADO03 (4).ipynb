**EJERCICIO 1.** Escribir una función que retorne los 5 usuarios que más veces accedieron al material de clases  durante la mañana, y los horarios en los que accedieron. Esta función recibe un parámetro opcional que indica si los nombres de usuarios se muestran en mayúsculas **("M")**, minúsculas **("m")** o como aparecen en el dataset **(None)**. Por defecto la función retorna los nombres tal como aparecen en el dataset. 

**NOTA:** la información sobre el material de clases se obtiene verificando el texto "Página: Material de las clases" en la columna "Contexto del evento" del dataset.

Consideramos que el horario de mañana es entre las 06:00 y las 12:00.

**EJERCICIO 2.** Utilizando la función anterior, escribir un programa que muestre dos listados correspondientes a la actividad  de los 5 usuarios que más veces accedieron al material de clases durante la mañana: uno en el cual  se muestren los nombres de usuario en minúsculas y otro tal cual como aparecen en el dataset.

## Solución

### EJERCICIO 1

```python
import csv
from pathlib import Path
from collections import Counter
from datetime import datetime

def top5_logs_matclase(csv_reader, mode=None):
    # Definir el rango horario
    horario_min = "06:00"
    horario_max = "12:00"

    # Filtrar los logs de "Material de las clases" y son en el horario mañana 
    logs = list(filter(lambda x: "Página: Material de las clases" in x["Contexto del evento"] and horario_min <= x["Hora"].split(",")[1].strip()[:5] <= horario_max, csv_reader))
    
    # Cuento los logs de los usuarios, luego los filtro por los que más logs tuvieron
    top_5 = Counter(map(lambda x: x["Nombre completo del usuario"], logs)).most_common(5)
    top_5_nombres = list(map(lambda x: x[0], top_5))

    # Armo un diccionario que guarda los usuarios con una lista de sus horas de acceso
    top5_usuarios_horas = {}
    for fila in logs:
        usuario = fila["Nombre completo del usuario"]
        hora = fila["Hora"].split(",")[1].strip()
        if usuario in top_5_nombres:
            if usuario not in top5_usuarios_horas:
                top5_usuarios_horas[usuario] = []
            top5_usuarios_horas[usuario].append(hora)

    # Modifico el formato del usuario en función del parámetro mode
    if mode == "M":
        top5_usuarios_horas = {usuario.upper(): horas for usuario, horas in top5_usuarios_horas.items()}
    elif mode == "m":
        top5_usuarios_horas = {usuario.lower(): horas for usuario, horas in top5_usuarios_horas.items()}

    return top5_usuarios_horas
```

```python
import csv
from pathlib import Path
from collections import Counter
from datetime import datetime

file_route = Path('archivos') / "logs_catedra.csv"

# Abro el archivo y lo paso itero con csv_readeMr
with open(file_route, 'r', encoding='utf-8') as archivo:
    csv_reader = csv.DictReader(archivo, delimiter=',')
    
    # Solicitar la entrada al usuario para el modo
    modo = input('Ingresa "M" para impresión del reporte en mayúsculas o "m" para impresión en minúsculas (Enter para mostrar como aparecen en tu dataset): ')
    
    # Llamo a la función que procesa los logs y obtengo el listado
    listado = top5_logs_matclase(csv_reader, mode=modo)

# Imprimo
for usuario, horas in listado.items():
    print(f"Horas de ingreso del usuario {usuario}")
    for hora in horas:
        print(hora)
```

```python
import pandas as pd
from pathlib import Path

# Ruta del archivo
file_route = Path('archivos') / "logs_catedra.csv"

def obtener_usuarios_mas_frecuentes(df, mayusculas=None):
    # Filtrar los registros donde "Contexto del evento" contiene "Página: Material de las clases"
    df_material = df[df['Contexto del evento'].str.contains('Página: Material de las clases', na=False)]
    
    # Convertir la columna 'Hora' a formato datetime para filtrado
    df_material['Hora'] = pd.to_datetime(df_material['Hora'], format='%d/%m/%y, %H:%M:%S')
    
    # Filtrar registros de la mañana (06:00 a 12:00)
    df_material_mañana = df_material[(df_material['Hora'].dt.hour >= 6) & (df_material['Hora'].dt.hour < 12)]
    
    # Contar los accesos por usuario
    accesos = df_material_mañana.groupby('Usuario afectado').size().reset_index(name='Accesos')
    
    # Ordenar por la cantidad de accesos, de mayor a menor
    accesos_ordenados = accesos.sort_values(by='Accesos', ascending=False).head(5)
    
    # Obtener los horarios de acceso de los 5 usuarios más frecuentes
    usuarios_top = []
    for usuario in accesos_ordenados['Usuario afectado']:
        horarios = df_material_mañana[df_material_mañana['Usuario afectado'] == usuario]['Hora'].dt.strftime('%H:%M:%S').tolist()
        usuarios_top.append((usuario, horarios))
    
    # Ajustar formato de los nombres de usuario según el parámetro 'mayusculas'
    if mayusculas == "M":
        usuarios_top = [(usuario[0].upper(), usuario[1]) for usuario in usuarios_top]
    elif mayusculas == "m":
        usuarios_top = [(usuario[0].lower(), usuario[1]) for usuario in usuarios_top]
    
    return usuarios_top


# Ejemplo de uso de la función
# Asegúrate de cargar el DataFrame desde el archivo CSV
df = pd.read_csv(file_route)

usuarios_men = obtener_usuarios_mas_frecuentes(df, mayusculas="m")
usuarios_original = obtener_usuarios_mas_frecuentes(df)

# Mostrar los resultados
print("Usuarios con nombres en minúsculas:")
for usuario in usuarios_men:
    print(f"Usuario: {usuario[0]}, Horarios: {usuario[1]}")

print("\nUsuarios con nombres originales:")
for usuario in usuarios_original:
    print(f"Usuario: {usuario[0]}, Horarios: {usuario[1]}")
```
